### Correlation between numeric features and price

```python
# Define the columns of interest for the correlation matrix
numeric_cols = ['accommodates', 'bedrooms', 'bathrooms', 'beds', 'price']

# Compute the correlation matrix from the DataFrame 'boyu_df' using the defined columns
corr_matrix = boyu_df[numeric_cols].corr()

# Create a new figure with specified size
plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix
# 'annot=True' adds labels to the cells
# 'cmap='coolwarm'' sets the color map of the heatmap to 'coolwarm'
# 'fmt=".2f"' formats the cell labels to 2 decimal places
# 'linewidths=0.5' sets the width of the lines that will divide each cell
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Set the title of the plot
plt.title('Correlation Matrix with Price')

# Display the plot
plt.show()
```

This is a correlation matrix heatmap that provides a visual representation of the correlation between different numerical variables in your dataset. The variables include ‘accommodates’, ‘bedrooms’, ‘bathrooms’, ‘beds’, and ‘price’.

Each cell in the heatmap represents the correlation coefficient between two variables. The correlation coefficient is a statistical measure that calculates the strength of the relationship between the relative movements of the two variables. The range of values for the correlation coefficient is -1.0 to 1.0. A correlation of -1.0 shows a perfect negative correlation, while a correlation of 1.0 shows a perfect positive correlation. A correlation of 0.0 shows no linear relationship between the movement of the two variables.

In this heatmap, darker colours represent a higher correlation. As you can see, ‘accommodates’, ‘bedrooms’, and ‘beds’ have a high positive correlation with ‘price’, indicating that as the number of accommodations, bedrooms, and beds increase, the price also tends to increase. On the other hand, ‘bathrooms’ have a moderate positive correlation with ‘price’.

### Setup

```python
import pycaret
# Checking the version of PyCaret
print(pycaret.__version__)
```

```python
# Importing all the modules from pycaret.regression
from pycaret.regression import *
```

Columns I will remove:

- “calculated_host_listings_count”: This column might not directly impact the rental price. The number of listings a host has does not necessarily affect the price of a particular accommodation.
  
- “latitude(North)”, “longitude(East)”: These are geographical coordinates. While location can affect price, using raw latitude and longitude might not be meaningful unless transformed into a more useful feature (like distance to the city centre).

- “amenities”: This column contains a list of amenities, which could be better handled by creating separate binary features for each amenity. Including it as is might not provide useful information to the model.

- “Unnamed: 25”: This column seems to be an artefact from the data loading process (possibly an extra column created due to a parsing error). It’s unlikely to contain useful information.

- “accommodates”: This column is highly correlated with “bedrooms” (0.81), which can lead to multicollinearity. To avoid this, one of these columns should be removed.

- “availability_30”: This feature has a weak correlation with the target variable ‘price’, and hence, might not contribute significantly to the predictive power of your model.
-
“review_scores_communication”, “review_scores_checkin”, “maximum_nights”, “review_scores_value”: These features have a weak correlation with the target variable ‘price’, and hence, might not contribute significantly to the predictive power of your model.

```python
# Initialize pycaret regression setup
# The setup function performs various preprocessing steps, including one-hot encoding for categorical features.
reg_setup = setup(data=boyu_df,                # Specify the DataFrame
                  target='price',              # Specify the target variable
                  ignore_features=["latitude(North)", "longitude(East)", 
                                   "amenities", "Unnamed: 25", 
                                   "availability_30",
                                   "calculated_host_listings_count",
                                   "review_scores_communication",
                                   "review_scores_checkin",
                                   "maximum_nights",
                                   "review_scores_value", "review_scores_location",
                                   "number_of_reviews"
                                   ],  # Ignore specified features
                  remove_outliers=True,         # Remove outliers
                  log_experiment=True,          # Log the experiment details
                  normalize=True,              # Normalize the data
                  transformation=True,         # Apply transformations
                  numeric_imputation='mean',   # Impute missing values in numeric columns
                  categorical_imputation='mode', # Impute missing values in categorical columns
                  session_id=120 
                 )
```

- Data: This is the dataset you’re using for the experiment, which is boyu_df.

- Target: This is the target variable that you’re trying to predict, which is price.

- Session ID: This is the seed for the random number generator used in the experiment, which is 123.

- Original Data: This shows the shape of the original dataset, which is (5834, 26), meaning there are 5834 rows and 26 columns.

- Preprocess: This is a boolean flag indicating whether preprocessing is applied to the dataset. It’s set to True, so preprocessing is applied.

- Imputation Type: This indicates the method used for imputing missing values. It’s set to simple.

- Numeric Imputation: This indicates the method used for imputing missing values in numeric columns. It’s set to mean, so missing values in numeric columns are replaced with the mean value of the column.

- Categorical Imputation: This indicates the method used for imputing missing values in categorical columns. It’s set to mode, so missing values in categorical columns are replaced with the mode (most frequent value) of the column.

- Fold Generator: This indicates the method used for generating folds for cross-validation. It’s set to KFold.

- Folds: This indicates the number of folds used for cross-validation, which is 10.

- CPU Jobs: This indicates the number of CPU cores used for parallel processing. It’s set to -1, which means all available cores are used.

- Use GPU: This is a boolean flag indicating whether a GPU is used for computations. It’s set to False, so a GPU is not used.

- Log Experiment: This is a boolean flag indicating whether the experiment is logged. It’s set to False, so the experiment is not logged.

- Experiment Name: This is the name of the experiment, which is reg-default-name.

- USI: This stands for Unique Session ID, which is 95c4 in this case.

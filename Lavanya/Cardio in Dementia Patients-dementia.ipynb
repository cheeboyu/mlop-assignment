# Data Cleaning

```python
import pandas as pd
Cdf = pd.read_csv ('cardio_data_processed.csv')
Cdf
```

Age shows the age of the patient in days while age_years shows the age of the patient in years. Thus, the age field can be removed as age_years will be used

```python
Cdf = Cdf.drop(['age'], axis=1)
Cdf
```

```python
Ddf = pd.read_csv ('health_dementia_data.csv')
Ddf
```

The 2 datasets will be linked up through the weight column. The dementia dataset's weight column will first be rounded up to a whole number to match the cardiovascular data's weight column.

```python
Ddf['Weight'] = Ddf['Weight'].round()
Ddf
```

```python
Ddf.rename(columns={'Weight': 'weight'}, inplace=True)
Ddf
```

```python
# Merge the datasets on the 'weight' column with a left join
combined_data = pd.merge(Ddf, Cdf[['id','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','cardio','age_years','bmi','bp_category','bp_category_encoded']], on='weight', how='left')
# Drop duplicates based on specific columns
combined_data = combined_data.drop_duplicates(subset=['Diabetic', 'AlcoholLevel', 'HeartRate', 'BloodOxygenLevel', 'BodyTemperature', 'weight', 'MRI_Delay', 'Dementia'])

# Drop rows with any missing values
combined_data = combined_data.dropna()
combined_data
```

```python
# Keep only data where Dementia is present as we are assuming all patients already have dementia
combined_data = combined_data[combined_data['Dementia'] == 1]
combined_data
```

```python
#Arrange the order
order = ['id', 'age_years', 'gender', 'weight', 'height', 'bmi', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'bp_category', 'bp_category_encoded', 'AlcoholLevel', 'HeartRate', 'BloodOxygenLevel', 'BodyTemperature', 'MRI_Delay', 'Diabetic', 'Dementia', 'cardio']

combined_data = combined_data[order]
combined_data
```

```python
#Renaming the Columns
column_mapping = {
    'age_years': 'Age',
    'gluc': 'Glucose',
    'alco': 'Alcohol',
    'cardio': 'Cardiovascular',
    'AlcoholLevel': 'Alcohol_level',
    'HeartRate': 'Heart_rate',
    'BloodOxygenLevel': 'Blood_oxygen_level',
    'BodyTemperature': 'Body_temp',
    'MRI_Delay': 'MRI_delay'
}

combined_data.rename(columns=column_mapping, inplace=True)

# Capitalize the first letter of all column names
combined_data.columns = [col.capitalize() for col in combined_data.columns]

combined_data
```

```python
# Drop unecessary columns: Alcohol as Alcohol Level is already present, Bp_category_encoded column as it contains the same values as Bp_category and Dementia column
# Drop the "Id" column as it should not be used as a factor affecting cardiovascular disease
unecessary = ['Alcohol','Bp_category_encoded', 'Id', 'Dementia']
combined_data.drop(columns=unecessary, inplace=True)
combined_data
```

```python
# Check for missing values in the entire DataFrame
missing_values = combined_data.isnull().sum()
print(missing_values[missing_values > 0])
```

Proactive risk assessment: Identifying and addressing missing values in data. By recognizing and addressing these missing values, the overall reliability and completeness of the dataset is enhanced. This also contributes to the effectiveness of the AI model, reducing the probability of inaccuracies.

```python
#Round columns with decimal points to 2 decimal places
combined_data = combined_data.round({'Bmi': 2, 'Alcohol_level': 2, 'Blood_oxygen_level': 2, 'Body_temp': 2, 'Mri_delay': 2})
combined_data
```

Prioritizing Ethical Considerations:
Ethical considerations are prioritized by rounding numerical columns to 2 decimal places.
This helps maintain consistency and fairness in numerical representations, reducing the potential impact of small variations and promoting transparency in reporting.

Ethical considerations are also prioritized by encoding categorical features with numerical values.
This ensures fairness and consistency in handling categorical data, reducing the risk of introducing biases associated with different category representations.

```python
#Encode Bp_category with numerical numbers: Normal 0, Elevated 1, Hypertension Stage 1 2, "Hypertension Stage 2 3
combined_data['Bp_category'] = combined_data['Bp_category'].str.replace('Normal','0')
combined_data['Bp_category'] = combined_data['Bp_category'].str.replace('Elevated','1')
combined_data['Bp_category'] = combined_data['Bp_category'].str.replace('Hypertension Stage 1','2')
combined_data['Bp_category'] = combined_data['Bp_category'].str.replace('Hypertension Stage 2','3')
combined_data
```

## Feature Scaling
Feature scaling is performed to ensure that all features have a similar scale, preventing some features from dominating others in terms of their influence on the machine learning model.Standard Scaling is used below where features are scaled to have a mean of 0 and a standard deviation of 1

```python
from sklearn.preprocessing import StandardScaler

# Select continuous columns
continuous_columns = ['Age', 'Weight', 'Height', 'Ap_hi', 'Ap_lo', 'Heart_rate', 'Blood_oxygen_level', 'Body_temp']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the selected columns
combined_data[continuous_columns] = scaler.fit_transform(combined_data[continuous_columns])
combined_data
```

## Feature Selection
Mitigating Biases and Ensuring Fairness: It is crucial to ensure that the features selected for the model do not introduce biases. Feature selection using ANOVA F-statistic helps mitigate biases by selecting features that have a significant impact on the target variable.
This statistical approach helps ensure that features chosen for the model are relevant and contribute meaningfully to predictions, reducing the risk of introducing biased information.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# 'X' is feature matrix and 'y' is target variable
X = combined_data.drop('Cardiovascular', axis=1)
y = combined_data['Cardiovascular']

# Using ANOVA F-statistic for feature selection
selector = SelectKBest(score_func=f_classif, k='all')
X_selected = selector.fit_transform(X, y)

# Get the selected features
selected_feature_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_feature_indices]

print("Selected Features:", selected_features)
```

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler

# Define the features and target variable
X = combined_data.drop('Cardiovascular', axis=1)
y = combined_data['Cardiovascular']

# Initialize the RandomOverSampler
oversampler = RandomOverSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Check the class distribution after resampling
resampled_class_distribution = pd.Series(y_resampled).value_counts()
print("Class Distribution after Resampling:")
print(resampled_class_distribution)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# Split the resampled data into training and testing sets
X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Convert the data types to float32
X_train_resampled = X_train_resampled.astype('float32')
X_test_resampled = X_test_resampled.astype('float32')
y_train_resampled = y_train_resampled.astype('float32')
y_test_resampled = y_test_resampled.astype('float32')

# Create a KNN model
model_knn_resampled = KNeighborsClassifier()

# Train the model on the resampled data
model_knn_resampled.fit(X_train_resampled, y_train_resampled)

# Predictions
y_pred_knn_resampled = model_knn_resampled.predict(X_test_resampled)

# Print classification report
print("KNN Classification Report on Resampled Data:")
print(classification_report(y_test_resampled, y_pred_knn_resampled))
```

```python
import joblib

# Specify the file path where you want to save the model
file_path = 'knn_model_resampled.pkl'

# Save the trained model to the specified file path
joblib.dump(model_knn_resampled, file_path)

print("Model saved successfully.")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Create a Logistic Regression model
model_lr = LogisticRegression(random_state=42)

# Train the model
model_lr.fit(X_train, y_train)

# Predictions
y_pred_lr = model_lr.predict(X_test)

# Print classification report
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
```

## Feedforward Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report

# Define and compile FFNN model
model_ffnn = Sequential()
model_ffnn.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model_ffnn.add(Dense(1, activation='sigmoid'))
model_ffnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train FFNN model
model_ffnn.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Predictions
y_pred_ffnn = model_ffnn.predict(X_test)
y_pred_binary_ffnn = (y_pred_ffnn > 0.5).astype(int)

# Print classification report
print("FFNN Classification Report:")
print(classification_report(y_test, y_pred_binary_ffnn))

```

## KNN

## Support Vector Machine with Linear kernel
